## Decision Tree Algorithm
 
 Decision Tree is a Supervised learning technique that can be used for both classification and Regression problems, but mostly it is preferred for solving Classification problems. It is a tree-structured classifier, where internal nodes represent the features of a dataset, branches represent the decision rules and each leaf node represents the outcome.
In a Decision tree, there are two nodes, which are the Decision Node and Leaf Node. Decision nodes are used to make any decision and have multiple branches, whereas Leaf nodes are the output of those decisions and do not contain any further branches.
The decisions or the test are performed on the basis of features of the given dataset.
It is a graphical representation for getting all the possible solutions to a problem/decision based on given conditions.
It is called a decision tree because, similar to a tree, it starts with the root node, which expands on further branches and constructs a tree-like structure.

 This would be done using the Python Scikit learn library.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
sns.set()
```

```python
iris= pd.read_csv(r'C:\Users\Ceejay\Downloads\Iris.csv')
iris
```

```python
# Getting to know our data
iris.info()
```

The dataset has 150 rows with six columns. All of the columns except Species are numerical columns.....

```python
# Checking for missing values
iris.isnull().sum()
```

```python
# Checking for duplicated values
iris.duplicated().sum()
```

There are no missing or duplicated values in the dataset.

```python
# Getting summary statistics of the dataset
iris.describe(include= 'all')
```

The above tables shows the summary statistics for all the columns in the dataset. There are three distinct species of the iris flower.

## Data Preparation

```python
# Dividing the data into attributes and labels
x= iris.drop(['Id', 'Species'], axis= 'columns')
y= iris.Species
```

```python
# Splitting the data into test set and training sets in the 80:20 ratio
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state= 207)
```

## Data Modelling

```python
# Instantiating the model and fitting on the training set
dt= DecisionTreeClassifier()
dt.fit(x_train, y_train)
print('Training completed')
```

## Lets visualize the Decision Tree to understand it better...

```python
text_representation= tree.export_text(dt)
print(text_representation)
```

The above is a text report showing the rules our model follows to generate predictions.

```python
plt.figure(figsize= [20, 15])
tree.plot_tree(dt, feature_names= y, filled= True)
plt.show()
```

The above is a visual repreentation of the rules in the dataset. When a data is fed into a model to be predicted, it follows this set of rules to give a prediction, 

Now we validate our models by making predictions on our test set

```python
y_pred= dt.predict(x_test)
# Comparing the actual values to the predicted values
df= pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
df
```

## Model Evaluation

Now we use a metric to measure the performance of our model. The metric of choice is the Accuracy metric. It specifies the amount of correct predictions made by our model ...

```python
score= accuracy_score(y_test, y_pred)
print('Accuracy score is {}%'.format(score.round(2)*100))
```

Now, we observe the general classificaton report for the predictions cross all species of iris flowers......

```python
print(classification_report(y_test, y_pred))
```

The model does very well in classifying the different iris species but predicts the iris-setosa very well. This can be seen in the f1-score.
